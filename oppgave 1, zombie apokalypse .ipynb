### FÃ¸rste utgave kode

```python
from pylab import*

mennesker = 500
k = 0.05 #reproduksjon
b = 1000 #bÃ¦reevne

#tid
t0 = 0 #tid (mÃ¥neder)
N = 10000
tid = 120
dt = tid/(N-1)

#matriser
t = zeros(N)
H = zeros(N)
Hder = zeros(N)

t[0] = t0
H[0] = mennesker

for i in range(N-1):
    Hder[i+1] = k*H[i]*(1 - H[i]/b)
    H[i+1] = H[i] + Hder[i]*dt
    t[i+1] = t[i]+ dt
    
plot(t,H)
title('vekst i menneskepopulasjon')
xlabel('antall mÃ¥neder')
ylabel('mennesker')
show()
```

```python
from pylab import*


overlevende = 500
zombier = 2000
k = 0.05 #reproduksjon
b = 1000 #bÃ¦reevne

#her kan du legge til en effektivitetsrate ut i fra forÃ¥tningsraten
#KAN LEGGE TIL EN MUTASJONSRATE
c = 0.000005 #menneske-zombie mÃ¸terate
d = 0.000005 #menneske-zombie mÃ¸terate 2

e = 0.0006

#tidssteg
t0 = 0 #tid (mÃ¥neder)
N = 10000
tid = 120
dt = tid/(N-1)


#matriser
t = zeros(N)
H = zeros(N)
Z = zeros(N)
Hder = zeros(N)
Zder = zeros(N)

#initierer matrisen
t[0] = t0
H[0] = overlevende
Z[0] = zombier

for i in range(N-1):
    Hder[i] = k*H[i]*(1 - H[i]/b) - c*H[i]*Z[i]
    Zder[i] = d*H[i]*Z[i] - e*Z[i]
    Z[i+1] = Z[i] + Zder[i]*dt
    H[i+1] = H[i] + Hder[i]*dt
    t[i+1] = t[i]+ dt
 
plot(t,H,'blue')
plot(t,Z,'green')
title('vekst i menneskepopulasjon')
xlabel('antall mÃ¥neder')
ylabel('mennesker')
grid()
show()
```
