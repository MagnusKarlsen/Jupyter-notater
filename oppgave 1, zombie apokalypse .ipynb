### Første utgave kode

```python
from pylab import*

mennesker = 500
k = 0.05 #reproduksjon
b = 1000 #bæreevne

#tid
t0 = 0 #tid (måneder)
N = 10000
tid = 120
dt = tid/(N-1)

#matriser
t = zeros(N)
H = zeros(N)
Hder = zeros(N)

t[0] = t0
H[0] = mennesker

for i in range(N-1):
    Hder[i+1] = k*H[i]*(1 - H[i]/b)
    H[i+1] = H[i] + Hder[i]*dt
    t[i+1] = t[i]+ dt
    
plot(t,H)
title('vekst i menneskepopulasjon')
xlabel('antall måneder')
ylabel('mennesker')
show()
```

### Andre utgave kode

```python
from pylab import*


overlevende = 500
zombier = 2000
k = 0.05 #reproduksjon
b = 1000 #bæreevne

#her kan du legge til en effektivitetsrate ut i fra foråtningsraten
#KAN LEGGE TIL EN MUTASJONSRATE
c = 0.000005 #menneske-zombie møterate
d = 0.000005 #menneske-zombie møterate 2

e = 0.0006

#tidssteg
t0 = 0 #tid (måneder)
N = 10000
tid = 120
dt = tid/(N-1)


#matriser
t = zeros(N)
H = zeros(N)
Z = zeros(N)
Hder = zeros(N)
Zder = zeros(N)

#initierer matrisen
t[0] = t0
H[0] = overlevende
Z[0] = zombier

for i in range(N-1):
    Hder[i] = k*H[i]*(1 - H[i]/b) - c*H[i]*Z[i]
    Zder[i] = d*H[i]*Z[i] - e*Z[i]
    Z[i+1] = Z[i] + Zder[i]*dt
    H[i+1] = H[i] + Hder[i]*dt
    t[i+1] = t[i]+ dt
 
plot(t,H,'blue')
plot(t,Z,'green')
title('vekst i menneskepopulasjon')
xlabel('antall måneder')
ylabel('mennesker')
grid()
show()
```

### Tredje utgave kode

```python
from pylab import*

overlevende = 500
zombier = 4500

#tidssteg
t0 = 0 #tid (måneder)
N = 10000
tid = 420
dt = tid/(N-1)
x = False

#matriser
t = zeros(N)
H = zeros(N)
Z = zeros(N)
Hder = zeros(N)
Zder = zeros(N)
E = zeros(N)

#initierer matrisen
t[0] = t0
H[0] = overlevende
Z[0] = zombier
E[0] = 0.0001


m = randint(10,20)
k = 0.05 #reproduksjon
b = 1700 #bæreevne
R = 0.0005 #råtningsrate
drapsrate_H = 0.0000075 #
drapsrate_Z = 0.00004 #

for i in range(N-1):
    
    if H[i] > 10:
        Hder[i] = k * H[i] * (1 - H[i]/b) - (E[i] + drapsrate_H) * Z[i]*H[i]
    elif H[i] <= 10:
        Hder[i+1] = 0
        
    
    if Z[i] > 0:
        Zder[i] = - R * Z[i] +(- drapsrate_Z + E[i]) * H[i] * Z[i]
    elif Z[i] <= 0:
        Zder[i+1] = 0
    
    Z[i+1] = Z[i] + Zder[i] * dt
    t[i+1] = t[i] + dt
    
    if H[i] > 10:
        H[i+1] = H[i] + Hder[i] * dt
    else:
        H[i+1] = 0
        R = 0.03
        
    t[i+1] = t[i] + dt
    
    if x:
        E[i+1] = E[i]
        drapsrate_H = 0.000006
    else: 
        E[i+1] = 0.000009
              
    if Z[i] == 0:
        b = 2500
        

plot(t,H,'blue')
plot(t,Z,'green')
title('vekst i menneske og zombie populsajon')
xlabel('antall måneder')
ylabel('mennesker')
grid()
show()
```

### Fjerde utgave kode 

```python
from pylab import*

overlevende = 500
zombier = 4500

#tidssteg
t0 = 0 #tid (måneder)
N = 10000
tid = 420
dt = tid/(N-1)
x = False

#matriser
t = zeros(N)
H = zeros(N)
Z = zeros(N)
Hder = zeros(N)
Zder = zeros(N)
E = zeros(N)

#initierer matrisen
t[0] = t0
H[0] = overlevende
Z[0] = zombier
E[0] = 0.0001


m = randint(10,20)
k = 0.05 #
b = 1700
e = 0.0005
drapsrate_H = 0.0000075 #
drapsrate_Z = 0.00004 #

for i in range(N-1):
    
    if H[i] > 10:
        Hder[i] = k * H[i] * (1 - H[i]/b) - (E[i] + drapsrate_H) * Z[i]*H[i]
    elif H[i] <= 10:
        Hder[i+1] = 0
        
    
    if Z[i] > 0:
        Zder[i] = - e * Z[i] +(- drapsrate_Z + E[i]) * H[i] * Z[i]
    elif Z[i] <= 0:
        Zder[i+1] = 0
    
    Z[i+1] = Z[i] + Zder[i] * dt
    t[i+1] = t[i] + dt
    
    if H[i] > 10:
        H[i+1] = H[i] + Hder[i] * dt
    else:
        H[i+1] = 0
        e = 0.03
    t[i+1] = t[i] + dt
    
    if 2 * round(Z[i],0) == round(H[i],0):
        E[i+1] = E[i] * m
        x = True
    elif x:
        E[i+1] = E[i]
        drapsrate_H = 0.000006
    else: 
        E[i+1] = 0.000009
              
    if Z[i] == 0:
        b = 5000
        

plot(t,H,'blue')
plot(t,Z,'green')
title('vekst i menneskepopulasjon')
xlabel('antall måneder')
ylabel('mennesker')
grid()
show()
```

# Endelig kode for zombie menneske utviklingen i Alexandria

```python
from pylab import*

#faktorer
mennesker = 500
zombier = 4500 #tenkte dette var et passende mengde zombier
mutasjonsfaktor = randint(10,20)
R_P_F = 0.05 #Reproduksjonsfaktor
bæreevne = 1700 #bæreevnen til menneskene, la ikke til noen bæreevne til zombiene siden de 
råtningsrate = 0.0005 #beskriver den naturlige råtningsraten til zombiene
drapsrate_H = 0.0000075 #beskriver raten av mennesker som drepes (spises) av zombier i løpet av en måned.
drapsrate_Z = 0.00004 #beskriver raten av zombier som blir drept av mennesker i løpet av en måned.


t0 = 0 #tid (måneder)
N = 10000 #antall intervaller
tid = 420 #antall måneder
dt = tid/(N-1)
x = False


#matriser
t = zeros(N)
H = zeros(N)
Z = zeros(N)
Hder = zeros(N)
Zder = zeros(N)
E = zeros(N)

#initierer matrisene
t[0] = t0
H[0] = mennesker
Z[0] = zombier
E[0] = 0.0001 #smittefaktor mellom zombier og mennesker, denne faktoren vil endre seg i løpet av koden ut i fra den tilfeldige mutasjonnen og ettersom det blir funnet en motgift


for i in range(N-1):
    
    #endring av stigning/synkning til menneske populasjonen i Alexandria
    if H[i] > 10:
        Hder[i] = R_P_F * H[i] * (1 - H[i]/bæreevne) - (E[i] + drapsrate_H) * Z[i]*H[i]
    elif H[i] <= 10:   #hindre at vi får negative verdier for menneskepopulasjonen
        Hder[i+1] = 0
        H[i+1] = 0
           
    #endring av stigningen/synkningen til zombie populasjonen i Alexandria  
    if Z[i] > 1:
        Zder[i] = - råtningsrate * Z[i] +(- drapsrate_Z + E[i]) * H[i] * Z[i]
    elif Z[i] <= 1:     #hindrer at vi får negative verdier for zombie populasjonen
        Zder[i+1] = 0
        Z[i+1] = 0
        bæreevne = 2500 #når zombiene har blitt utryddet vil bæreevnen til området øke siden mer areal er frigjort
        
        
        
    #endring av Zombie populasjon per tidsintervall
    Z[i+1] = Z[i] + Zder[i] * dt
    
    #endring av Menneske populasjon per tidsintervall
    if H[i] > 10:             #hvis populasjonen blir for lav vil ikke landsbyen klare seg
        H[i+1] = H[i] + Hder[i] * dt
    else:                     #når menneskepopulasjonen synker for mye vil populasjonnen bli null dette gjør jeg siden menneske bestanden noen ganger i ganges- 
                              #-med Z[i] i mattestykkene som hindrer zombie populasjone fra å synke som den skal uten mennesker.
        
        H[i+1] = 0
        råtningsrate = 0.03   #dette er råtningraten hvis menneskene utryddes siden disse zombiene foråtner fortere hvis de ikke får konsumert mennesker
        
    #endring av tid per tidsintervall
    t[i+1] = t[i] + dt
    
    
    
    #denne er annsvarlig for en mutasjon som forekommer når det er dobbelt så mange mennesker som zombier da vil en mutasjon forekomme som endrer smitteraten drastisk
    if 2 * round(Z[i],0) == round(H[i],0) and i < 4800: 
        E[i+1] = E[i] * mutasjonsfaktor
        x = True
        y = i
    elif x:
        E[i+1] = E[i]        #sørger for en konstant vekst
        drapsrate_H = 0.000006  #drapsraten minsker fordi zombiene heller smitter offeret
    else: 
        E[i+1] = 0.000009    #vanlig smitterate mellom mennesker og zombier
        
        
    #bestemmer hvorvidt menneskene i Alexandria vil klare å finne en motgift for zombieviruset i forhold til mutasjonsfaktoren    
    if H[4760] > 170:        #de må ha en hvis mengde folk for å finne fram til en motgift
        råtningsrate = 0.03  #zombine vil foråtne fortere med motgiften
        E[i+1] = 0        #de vil ikke lenger være i stand til å smitte noen
        svar = 'menneskene i alexandria vil klare seg!'
    else:
        svar = 'menneskene i Alexandria er fortapte!'

        
#svar        
print('det er',round(H[4760]), 'mennesker igjen ved år 17')         
print(svar)
 
    
#plotting av grafene
plot(t[y],Z[y],'ro')         #plotter det kritiske punktet for mutasjonen til zombiene
plot(t[4760],H[4760],'yo')   #plotter det kritiske punktet for alexandrias landsby sin overlevelse

plot(t,H,'-b', label = 'Mennesker')
plot(t,Z,'-g',label = 'Zombier')

data = plot(t,H,'-b', label = 'Mennesker') + plot(t,Z,'-g', label = 'Zombier')
datatittel = [i.get_label() for i in data]
legend(data,datatittel)

title('zombie- og menneske- popusjonene i Alexandria')
xlabel('antall måneder')
ylabel('antall mennesker og zombier')
grid()
show()
```
