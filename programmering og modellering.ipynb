## lesing og plotting av data

```python
from pylab import*

data = loadtxt("test.text", skiprows = 1, delimiter =',')

månde = data[:,0]
solflekker = data[:,1]

plot(månde,solflekker)


snitt = mean(solflekker)
avvik = std(solflekker)

print("snitt:", snitt, "standardavvik:", avvik)
```

loadtxt() er en funksjon en annvender for å henvise til et annet dokument for å hente ut informasjon fra dette dokumentet. 

skiprows brukes til å fortelle hvilke linjer fra dokumentet som skal ses bort i fra.

delimiter brukes til deffinere hva som skiller innformasjonen fra hverandre.

funksjonen mean() finner gjennomsnittet av tallen i en array
std() finner gjennomsnittet av avviket fra gjennomsnittet.

# Regresjon
regresjon går ut på at en finner en passennde funskjon som beskriver en trend for en rekke pukter

```python

from pylab import*

data = loadtxt('tid og possisjon heis', skiprows = 1, delimiter ='	')
tid = data[:,0]
strekning = data[:,1]

scatter(tid,strekning)

grad = 1

reg_strekning = polyfit(tid,strekning,grad)

x = linspace(0,55,100000)
a = reg_strekning[0]
b = reg_strekning[1]


y = a*x + b

scatter(tid,strekning)
plot(x,y,color ='limegreen')
show()
```

først henviser vi til et annet dokument med innformasjon. deretter deffinnere vi tid og strekning ut i fra om det er nullte, eller første element.

så deffinner vi om hvilken grads polynomfunksjon vi ønsker
polyfit brukes for å finne passende konstanter for den ønskede funksjonen.
disse setter vi så inn i funksjonen y = ax + b

samtidig bruker vi scatter for at alle de tidligere puktene skal settes inn med vår nye funksjon.
